The 2nd round of fixtures are the reverse of the 1st round, i.e. switch the home and away teams. Increment matchweek ID by 9 for the 2nd round

Create a list of dates across the season where there is no fixture due to cup competitions, winter break etc.

```python
from datetime import date
fixture_days = "Mon"
prem_exclude = [date(2023, 10, 23), date(2023, 12, 4), date(2023, 12, 11), date(2023, 12, 18), date(2023, 12, 25),
                date(2024, 1, 1), date(2024, 2, 5)]
prem_exclude
```

```python
div1_exclude = [date(2023, 10, 23), date(2023, 12, 4), date(2023, 12, 25), date(2024, 1, 1), date(2024, 2, 5)]
div1_exclude
```

Generate a list of dates for each matchweek excluding the above dates

```python
prem_matchweeks = pd.DataFrame(
{
    "matchweek" : list(range(1, 19)),
    "Date dd/mm/yyyy": pd.bdate_range(date(2023, 9, 25), date(2024, 3, 11), freq = 'C', weekmask = fixture_days,
                                      holidays = prem_exclude)
})
prem_matchweeks["Time HH:MM"] = "19:00"
prem_matchweeks["Date dd/mm/yyyy"] = prem_matchweeks["Date dd/mm/yyyy"].dt.strftime("%d/%m/%Y")
prem_matchweeks.head(5)
```

```python
div1_matchweeks = pd.DataFrame(
{
    "matchweek" : list(range(1, 23)),
    "Date dd/mm/yyyy": pd.bdate_range(date(2023, 9, 25), date(2024, 3, 25), freq = 'C', weekmask = fixture_days,
                                      holidays = div1_exclude)
})
div1_matchweeks["Time HH:MM"] = "19:00"
div1_matchweeks["Date dd/mm/yyyy"] = div1_matchweeks["Date dd/mm/yyyy"].dt.strftime("%d/%m/%Y")
div1_matchweeks.head(5)
```

Create dataframes of Premier Division and Division 1 teams

```python
prem_teams = pd.DataFrame(
{
    "team_letter" : ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'],
    "team_code" : ["SMA", "SMD", "SMC", "SME", "PSA", "PSB", "CCA", "CCC", "SMB", "BCJ"]
})
prem_teams.head(5)
```

```python
div1_teams = pd.DataFrame(
{
    "team_letter" : ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l'],
    "team_code" : ["MCA", "MCB", "MCC", "HDS", "SSA", "SSC", "CHA", "CHB", "PL", "PC", "SSB", "BYE"]
})
div1_teams.head(5)
```

Merge the datasets to create a Premier Division fixture dataframe and a CSV for uploading to LeagueRepublic

```python
prem_fixtures = pd.merge(pd.merge(pd.merge(prem_fixtures, prem_teams, how = "inner", left_on = "home",
                                           right_on = "team_letter")
                 .rename(columns = {"team_code": "Home Team"}),
                prem_teams, how = "inner", left_on = "away",
                                  right_on = "team_letter").rename(columns = {"team_code": "Away Team"})
                , prem_matchweeks, how = "inner", on = "matchweek")
prem_fixtures["Division"] = "Premier Division"
prem_fixtures["Venue"] = ""
prem_fixtures["Pitch"] = ""
prem_fixtures["Home Score"] = ""
prem_fixtures["Away Score"] = ""
prem_fixtures = prem_fixtures.sort_values("matchweek")[["Date dd/mm/yyyy", "Time HH:MM", "Division", "Home Team", "Away Team",
                               "Venue", "Pitch", "Home Score", "Away Score"]]
prem_fixtures.to_csv("Premier fixtures 23-24.csv", index=False)
prem_fixtures.head(5)
```

Merge the datasets to create fixtures for Division 1 and export to CSV for uploading to the website

```python
div1_fixtures = pd.merge(pd.merge(pd.merge(div1_fixtures, div1_teams, how = "inner", left_on = "home",
                                           right_on = "team_letter")
                 .rename(columns = {"team_code": "Home Team"}),
                div1_teams, how = "inner", left_on = "away",
                                  right_on = "team_letter").rename(columns = {"team_code": "Away Team"})
                , div1_matchweeks, how = "inner", on = "matchweek")
div1_fixtures = div1_fixtures.loc[div1_fixtures["Home Team"] != "BYE"]
div1_fixtures = div1_fixtures.loc[div1_fixtures["Away Team"] != "BYE"]
div1_fixtures["Division"] = "Division 1"
div1_fixtures["Venue"] = ""
div1_fixtures["Pitch"] = ""
div1_fixtures["Home Score"] = ""
div1_fixtures["Away Score"] = ""
div1_fixtures = div1_fixtures.sort_values("matchweek")[["Date dd/mm/yyyy", "Time HH:MM", "Division", "Home Team", "Away Team",
                               "Venue", "Pitch", "Home Score", "Away Score"]]
div1_fixtures.to_csv("Div 1 fixtures 23-24.csv", index=False)
div1_fixtures.head(5)
```
